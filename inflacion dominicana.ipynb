```python
import pandas as pd

datos = pd.read_excel("inflacion_dominicana.xlsx")
```

```python
datos.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 9))
plt.plot(datos['year'], datos['Annual average inflation (consumer prices) rate'],
         marker='^', linestyle='-.')
plt.xlabel('Año')
plt.ylabel('Promedio anual de inflación anual ')
plt.title('Rango de infación anual desde 1980')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
X = datos['year']
y = datos['Annual average inflation (consumer prices) rate']
```

```python
y
```

```python
X.values
```

```python
X = X.values.reshape(-1, 1)
y = y.values.reshape(-1, 1)
```

```python
modelo.fit(X,y )
```

```python
anio = int(input("Ingrese el año para predecir la inflación:\n "))

predicciones = modelo.predict([[anio]])
print(f"La predicción de la inflación promedio para el año {anio} es: {predicciones[0]}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(datos['year'], datos['Annual average inflation (consumer prices) rate'],
          marker='*', linestyle='-.')
plt.plot(datos['year'], modelo.predict(X), color='red')
plt.xlabel('Año')
plt.ylabel('Promedio anual de inflación anual')
plt.title('Rango de infación anual desde 1980 - 2024')
plt.grid(True)
plt.show()
```

```python
import math
exponencial = math.exp(predicciones[0])
print(f"La predicción de la inflación promedio para el año {datos['Annual average inflation (consumer prices) rate']} es: {exponencial}")
```

```python
plt.figure(figsize=(20, 6))
plt.plot(datos['year'], datos['Annual average inflation (consumer prices) rate'],
          marker='*', linestyle='-.')
plt.plot(datos['year'], modelo.predict(X), color='red')
plt.plot([[2025]], exponencial, marker='*', linestyle='--', markersize=10, color='green')
plt.xlabel('Año')
plt.ylabel('Promedio anual de inflación anual')
plt.title('Rango de infación anual en promedio desde 1980 - 2024')
plt.grid(True)
plt.show()
```
